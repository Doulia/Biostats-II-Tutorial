# SAS Code Tutorial 1

```sas
proc import datafile="tutorial1.csv"
        out=tutorial1
        dbms=csv
        replace;
run;
```

```sas
proc print data=tutorial1;
run;
```

```sas
data df_modified; 
    set tutorial1;
    drop VAR1; 
run; 
```

```sas
proc print data=df_modified  (obs=5);
run;
```

```sas
proc glm data=df_modified plot=diagnostics;
   model sales = youtube facebook / tolerance;
run; 
```

```sas
proc glm data=df_modified plot=diagnostics;
   model sales = youtube|facebook / tolerance;
run; 
```
